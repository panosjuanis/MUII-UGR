### Exercise 3

Decoder

```python
#Read and clean data
decoder_data = ""
with open('decoder_data.txt') as f:
    for line in f:
        decoder_data += line.strip()

decoder_data = [*decoder_data]
decoder_data = np.array(decoder_data)
valid_idx = np.where((decoder_data == '0') | (decoder_data == '1') | (decoder_data == '2'))
decoder_data = decoder_data[valid_idx]
decoder_data = decoder_data.astype(int)
print("Input values read from the decoder: ")


# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = '0', '1', '2'
sizes = [np.sum(decoder_data == 0), np.sum(decoder_data == 1), np.sum(decoder_data == 2)]
explode = (0, 0.1, 0.1)  # only "explode" the 2nd slice

fig1, ax1 = plt.subplots()
plt.title("Decoder data, n = "  + str(decoder_data.size))
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()


#Histogram
plt.hist(decoder_data)
plt.show() 
```

### Exercise 4

Healthcare data

```python
#Read data
categories = ['Tabaquismo',  'Obesidad', 'Tension', 'Pulsaciones', 'Edad', 'Hierro']
patient_data = np.array([[0.34, 0.2, 0.56, 0.12, 0.9, 0.8],
                     [0.5, 0.8, 0.9, 0.7, 0.5, 0.9],
                     [0.12, 0.15, 0.23, 0.45, 0.6, 0.1],
                     [0.1, 0.6, 0.1, 0.4, 0.3, 0.9]])
df = pd.DataFrame(data=patient_data, columns=categories)
#Plot
import plotly.express as px
fig = px.parallel_coordinates(df, color="Tabaquismo",
                              dimensions=categories,
                              color_continuous_scale=px.colors.diverging.Tealrose,
                              color_continuous_midpoint=2)
fig.show()
```
